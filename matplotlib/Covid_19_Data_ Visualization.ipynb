```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# plt.style.use('classic')
sns.set_theme(style="white")
```

### Import data set

```python
Covid_19=pd.read_csv('covid19_merged.csv')
Covid_19.head()
```

### See the nulll values using heat map

```python
plt.figure(figsize=(20,10))
sns.heatmap(Covid_19.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("visualizing_missing_data_with_heatmap_Seaborn_Python.png", dpi=100)
```

```python
#Dis plot
plt.figure(figsize=(20,10))
sns.displot(
    data=Covid_19.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
plt.savefig("visualizing_missing_data_with_barplot_Seaborn_distplot.png", dpi=100)
# plt.show()
```

```python
#Sum the Total number of null values
Covid_19.isnull().stack().sum()
```

```python
#Get the Total number of colunm
Covid_19.columns
```

```python
Covid_19.info()
```

```python
#Get the useful colunm
covid_19=Covid_19[['country', 'covid_confirmed_4_28_20',
       'covid_deaths_4_28_20', 'covid_recovered_4_28_20', 'covid19_first_date','Median_age','gdp_usd_million','sex_male_to_female_total','code_3digit_x','latitude','longitude']]
```

```python
#Now again check value
covid_19.isnull().sum()
```

```python
#Check info of Data set
import numpy as np
covid_19.info()
```

```python
#Convert the object colunm to Float/int
#Median_age
Covid_19['Median_age'] = Covid_19['Median_age'].replace(['N.A.'], np.nan)
covid_19['Median_age']=Covid_19['Median_age'].astype(float, errors = 'raise')
Value=Covid_19['Median_age'].median()
covid_19['Median_age']=covid_19['Median_age'].fillna(value=Value)
covid_19['Median_age'].isnull().sum()
```

```python
# GDP_USD_doller
covid_19['gdp_usd_million']=covid_19['gdp_usd_million'].fillna(value=Covid_19['gdp_usd_million'].median())
covid_19['gdp_usd_million'].isnull().sum()
```

```python
Covid_19['sex_male_to_female_total'] =Covid_19['sex_male_to_female_total'].replace(['-'], np.nan)
Covid_19['sex_male_to_female_total']=Covid_19['sex_male_to_female_total'].astype(float, errors = 'raise')
covid_19['sex_male_to_female_total']=Covid_19['sex_male_to_female_total'].fillna(value=Covid_19['sex_male_to_female_total'].median())
covid_19['sex_male_to_female_total'].isnull().sum()
```

```python
#'code_3digit_x'
covid_19.drop(index=covid_19.index[44], axis=0, inplace=True)
covid_19['code_3digit_x'].isnull().sum()
```

```python
# latitud and longitude
covid_19['latitude']=covid_19['latitude'].fillna(value=covid_19['latitude'].mean())
covid_19['longitude']=covid_19['longitude'].fillna(value=covid_19['longitude'].mean())
covid_19['latitude'].isnull().sum()
```

```python
# Now again check the null values
covid_19.isnull().sum()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(covid_19.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("visualizing_missing_data_with_heatmap_Seaborn_Python.png", dpi=100)
```

### Analyzing of Spreading of Covid 19 in all over the world

```python
#Convert the covid first date colunm from obj into d.t
```

```python
covid_19['covid19_first_date'] = pd.to_datetime(covid_19['covid19_first_date']).dt.date
```

#### Number of days that Covid spread in all over the world

```python
covid_19['covid19_first_date'].max()-covid_19['covid19_first_date'].min()
```

### Dervied Spread time colunm from the first date


```python
covid_19['Spread_time']=covid_19['covid19_first_date']-covid_19['covid19_first_date'].min()
covid_19['Spread_time'].head()
```

```python
#Top 20 country where covid is first spread
Top_20_spread=covid_19.nsmallest(20, ['Spread_time'])
Top_20_spread=Top_20_spread[['country','Spread_time']]
Top_20_spread
```

```python
#Spread Time and number of countries
s=covid_19.groupby([covid_19['Spread_time']])["country"].count()
```

```python
i=np.arange(0,45)
len(i)
values=s.values
len(values)
averge=values.mean()
avergee=np.full(shape=len(i),fill_value=averge)
```

```python
#How to Spread the Covid per day in all over the world
fig, ax = plt.subplots(figsize=(12, 7))
# x = np.linspace(0, 20, 1000)
ax.plot(i, values,':o',label='Covid_spread/per day')
plt.plot(i,avergee,label='average of per day of covid spread')
ax.axis('equal')
plt.title('Spreading of Covid 19 in all over the world',fontsize=15)
plt.xlabel('Number of Days', fontsize=16)
plt.ylabel('Number of Countries', fontsize=16)
plt.legend(loc='best',shadow=True) #loc=upper left 

ax.annotate('average of every day spreading covid 19', xy=(6.28,4 ), xytext=(10,10),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('Spreading of Covid_19 per day', xy=(20,4), xytext=(2, -6),
           arrowprops=dict(facecolor='black', shrink=0.05));
plt.savefig("Spreading of covid_19.png", dpi=500)
plt.show()
```
